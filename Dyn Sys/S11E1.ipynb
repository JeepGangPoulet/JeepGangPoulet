```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import root
```

```python
rs=1
def xdot(x):
    r=rs
    return r-x-np.exp(-x)
xs = np.linspace(-1.5,3,1000)
x_dots = xdot(xs)
x0 = root(xdot,-0.01)
fig1 = plt.figure(1)
plt.title(f'$xdot$ for r={rs}')
plt.xlabel('x')
plt.ylabel(r'$\dot{x}$')
plt.plot(xs,x_dots)
plt.scatter(x0.x[0],0,color='r',s=50,label=f'$x_0$={np.round(x0.x[0],3)}')
plt.legend()
plt.grid()
```
